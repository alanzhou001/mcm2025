# 导入所需库
导入pandas和numpy库。

```python
import pandas as pd
import numpy as np
```

# 读取数据
使用pandas的read_csv函数读取CSV文件：summerOly_athletes.csv、summerOly_hosts.csv、summerOly_medal_counts.csv、summerOly_programs.csv。

```python
# 读取数据

# 使用pandas的read_csv函数读取CSV文件
athletes_df = pd.read_csv(r'D:\mcm2025\data\2025_Problem_C_Data\2025_Problem_C_Data\summerOly_athletes.csv', encoding='ISO-8859-1')
hosts_df = pd.read_csv(r'D:\mcm2025\data\2025_Problem_C_Data\2025_Problem_C_Data\summerOly_hosts.csv', encoding='ISO-8859-1')
medal_counts_df = pd.read_csv(r'D:\mcm2025\data\2025_Problem_C_Data\2025_Problem_C_Data\summerOly_medal_counts.csv', encoding='ISO-8859-1')
programs_df = pd.read_csv(r'D:\mcm2025\data\2025_Problem_C_Data\2025_Problem_C_Data\summerOly_programs.csv', encoding='ISO-8859-1')

# 显示每个数据框的前5行数据
print(athletes_df.head())
print(hosts_df.head())
print(medal_counts_df.head())
print(programs_df.head())
```

```python
# 查看数据基本信息

# 使用head()函数查看每个数据集的前5行数据
print(athletes_df.head())
print(hosts_df.head())
print(medal_counts_df.head())
print(programs_df.head())

# 使用info()函数查看每个数据集的基本信息
print(athletes_df.info())
print(hosts_df.info())
print(medal_counts_df.info())
print(programs_df.info())

# 使用describe()函数查看每个数据集的统计信息
print(athletes_df.describe())
print(hosts_df.describe())
print(medal_counts_df.describe())
print(programs_df.describe())
```

# 检查缺失值
使用isnull().sum()函数检查每个数据集中的缺失值情况。

```python
# 检查缺失值

# 使用isnull().sum()函数检查每个数据集中的缺失值情况
print("athletes_df 缺失值情况：")
print(athletes_df.isnull().sum())

print("\nhosts_df 缺失值情况：")
print(hosts_df.isnull().sum())

print("\nmedal_counts_df 缺失值情况：")
print(medal_counts_df.isnull().sum())

print("\nprograms_df 缺失值情况：")
print(programs_df.isnull().sum())
```

# 检查重复值
使用duplicated().sum()函数检查每个数据集中的重复值情况。

```python
# 检查重复值

# 使用duplicated().sum()函数检查每个数据集中的重复值情况
print("athletes_df 重复值情况：")
print(athletes_df.duplicated().sum())

print("\nhosts_df 重复值情况：")
print(hosts_df.duplicated().sum())

print("\nmedal_counts_df 重复值情况：")
print(medal_counts_df.duplicated().sum())

print("\nprograms_df 重复值情况：")
print(programs_df.duplicated().sum())
```

# 数据清洗建议
根据缺失值和重复值的检查结果，提出数据清洗的建议。

```python
# 数据清洗建议

# 根据缺失值和重复值的检查结果，提出数据清洗的建议

# 检查 athletes_df 的缺失值和重复值情况
athletes_missing = athletes_df.isnull().sum()
athletes_duplicates = athletes_df.duplicated().sum()

# 检查 hosts_df 的缺失值和重复值情况
hosts_missing = hosts_df.isnull().sum()
hosts_duplicates = hosts_df.duplicated().sum()

# 检查 medal_counts_df 的缺失值和重复值情况
medal_counts_missing = medal_counts_df.isnull().sum()
medal_counts_duplicates = medal_counts_df.duplicated().sum()

# 检查 programs_df 的缺失值和重复值情况
programs_missing = programs_df.isnull().sum()
programs_duplicates = programs_df.duplicated().sum()

# 打印数据清洗建议
print("数据清洗建议：\n")

# athletes_df 清洗建议
if athletes_missing.any() or athletes_duplicates > 0:
    print("athletes_df 清洗建议：")
    if athletes_missing.any():
        print("- 处理缺失值：")
        print(athletes_missing[athletes_missing > 0])
    if athletes_duplicates > 0:
        print("- 删除重复值")
    print()

# hosts_df 清洗建议
if hosts_missing.any() or hosts_duplicates > 0:
    print("hosts_df 清洗建议：")
    if hosts_missing.any():
        print("- 处理缺失值：")
        print(hosts_missing[hosts_missing > 0])
    if hosts_duplicates > 0:
        print("- 删除重复值")
    print()

# medal_counts_df 清洗建议
if medal_counts_missing.any() or medal_counts_duplicates > 0:
    print("medal_counts_df 清洗建议：")
    if medal_counts_missing.any():
        print("- 处理缺失值：")
        print(medal_counts_missing[medal_counts_missing > 0])
    if medal_counts_duplicates > 0:
        print("- 删除重复值")
    print()

# programs_df 清洗建议
if programs_missing.any() or programs_duplicates > 0:
    print("programs_df 清洗建议：")
    if programs_missing.any():
        print("- 处理缺失值：")
        print(programs_missing[programs_missing > 0])
    if programs_duplicates > 0:
        print("- 删除重复值")
    print()
```

## 处理缺失数据

```python
# 使用0填充所有缺失值
programs_df.fillna(0, inplace=True)

# 检查清洗后的缺失值
print("\nPrograms DataFrame Missing Values After Cleaning:")
print(programs_df.isnull().sum())
```
